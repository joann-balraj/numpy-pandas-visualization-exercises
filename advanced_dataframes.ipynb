### 10. What is the oldest date in the to_date column?



```python
url = get_db_url("employees")
sql = '''
SELECT
    to_date
FROM titles
ORDER BY to_date ASC
'''

titles = pd.read_sql(sql, url)
titles.head(1)
```

### 11. What is the most recent date in the to_date column?

```python
url = get_db_url("employees")
sql = '''
SELECT
    to_date
FROM titles
ORDER BY to_date DESC
'''

titles = pd.read_sql(sql, url)
titles.head(1)
```
