* Title your chart "Big O Notation"
* Label your x axis "Elements"
* Label your y axis "Operations"
* Label your curves or make a legend for the curves
* Use LaTex notation where possible
* Curves to graph

1. y = On + 1 , and label the curve "O(1)"

2. y = log(n) , and label the curve "O(log n)"

3. y = n , and label the curve "O(n)"

4. y = n * log(n) , and label it "O(n log n)"

5. y = n^2 , and label it "O(n^2)"

6. y = 2^n , and label it "O(2^n)"

7. y = n! , and label it "O(n!)"

8. y = n^n , and label it "O(n^n)"


```python
import matplotlib.pyplot as plt
import math
import numpy as np
from random import randint
import random
import pandas as pd
```

```python
n = [1,2,3,4,5]
y1 = [(0 * n + 1) for n in n]
y2 = [math.log(n) for n in n]
y3 = [n for n in n]
y4 = [(n * math.log(n)) for n in n]
y5 = [(n ** 2) for n in n]
y6 = [(2 ** n) for n in n]
y7 = [(math.factorial(n)) for n in n]
y8 = [(n ** n) for n in n]


plt.figure(figsize = (20,40))
plt.suptitle('Big O Notation')
plt.subplot(4,2,1)
plt.plot(n,y1, color = 'k', marker = '*',
         linestyle = '-', label ='1. y = On + 1')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()



plt.subplot(4,2,2)
plt.plot(n,y2, color = 'k', marker = '*',
         linestyle = '-', label ='2. y = O(log n)')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()



plt.subplot(4,2,3)
plt.plot(n,y3, color = 'k', marker = '*',
         linestyle = '-', label ='3. y = O(n)')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()




plt.subplot(4,2,4)
plt.plot(n,y4, color = 'k', marker = '*',
         linestyle = '-', label ='4. y = O(n log n)')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()



plt.subplot(4,2,5)
plt.plot(n,y5, color = 'k', marker = '*',
         linestyle = '-', label ='5. y = O(n^2)')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()


plt.subplot(4,2,6)
plt.plot(n,y6, color = 'k', marker = '*',
         linestyle = '-', label = '6. y = O(2^n)')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()


plt.subplot(4,2,7)
plt.plot(n,y7, color = 'k', marker = '*',
         linestyle = '-', label = '7. y = O(n!)')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()


plt.subplot(4,2,8)
plt.plot(n,y8, color = 'k', marker = '*',
         linestyle = '-', label = '8. y = O(n^n)')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend();
```
