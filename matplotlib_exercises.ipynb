### 1. Use matplotlib to plot the following equation:

`y = x2 − x + 2`

You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

```python
import matplotlib.pyplot as plt
x = range(-5, 7)
y = [(n ** 2 - n + 2) for n in x]

plt.plot(x,y)
plt.annotate('origin', 
             xy = (0, 2), 
             xytext = (1, 10),
             arrowprops = dict(facecolor = 'black'));
```

### 2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

1. y=√x

2. y=x^3

3. y=2^x

4. y=1/(x+1)

```python
import math
import numpy as np
from random import randint
import random
import pandas as pd
```

```python
# 1.
x1 = [x1 * 1 for x1 in range(0,70)]
y1 = [math.sqrt(x1) for x1 in x1]
plt.plot(x1,y1)
plt.title('$\sqrt{x}$');
```

```python
#2.
x2 = np.linspace(-5,5,100)
y2 = x2 ** 3
plt.plot(x2,y2)
plt.title('$x^{3}$');
```

```python
#3.
x3 = np.linspace(-5,5,100)
y3 = 2**x3
plt.plot(x3, y3)
plt.title('$2^{x}$');
```

```python
#4.
x4 = np.linspace(-5,5,100)
y4 = 1/(x4+1)
plt.plot(x4,y4);
```

### 3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
x1 = [x1 * 1 for x1 in range(0,70)]
y1 = [math.sqrt(x1) for x1 in x1]
x2 = np.linspace(-5,5,100)
y2 = x2 ** 3
x3 = np.linspace(-5,5,100)
y3 = 2**x3
x4 = np.linspace(-5,5,100)
y4 = 1/(x4+1)


plt.figure(figsize = (10,8))


plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.title('1. y=√x')

plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.title('2. y=x^3')

plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.title('3. y=2^x')

plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.title('4. y=1/(x+1)');
```

### 4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
x1 = [x1 * 1 for x1 in range(0,70)]
y1 = [math.sqrt(x1) for x1 in x1]
x2 = np.linspace(-5,5,100)
y2 = x2 ** 3
x3 = np.linspace(-5,5,100)
y3 = 2**x3
x4 = np.linspace(-5,5,100)
y4 = 1/(x4+1)


plt.figure(figsize = (10,8))


plt.subplot(2,2,1)
plt.plot(x1,y1,color = 'blue', marker = '.',
         linestyle = '-', label = '1. y=√x')

plt.subplot(2,2,2)
plt.plot(x2,y2,color = 'r', marker = '',
         linestyle = '-', label = '2. y=x^3')

plt.subplot(2,2,3)
plt.plot(x3,y3,color = 'k', marker = '*',
         linestyle = '-', label = '3. y=2^x')

plt.subplot(2,2,4)
plt.plot(x4,y4,color = 'm', marker = ',',
         linestyle = '-', label = '4. y=1/(x+1)')





# add label for x and y axis
plt.xlabel('X')
plt.ylabel('Y')
# add title
plt.title('4 equations')
# add gridlines
plt.grid('True', ls = '--')
# add legend
plt.legend();
```
